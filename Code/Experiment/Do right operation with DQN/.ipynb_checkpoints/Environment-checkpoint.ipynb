## Import Package

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import random
```

## Environment

```python
class Environment():
    
    def __init__(self, stock1="GM", stock2="F"):
        # get two stocks' close price
        self.stock1_price = self.get_close_price(filename=stock1)
        self.stock2_price = self.get_close_price(filename=stock2)
        self.spread_ratio = None
        
        # number of units of stock1 and stock2 which agent have
        self.agent_stock1 = 100.0
        self.agent_stock2 = 100.0

        # current time interval
        self.cur = None
        
        # genearte label: label for spread ratio
        self.dataset = self.gen_dataset()
        
        
        
        
    def get_close_price(self, filename):
        stock_df = pd.read_csv("../../../Data/{}.csv".format(filename))
        return stock_df["Close"].values[-2400:-1]

        
        
    def gen_dataset(self):
        '''
        self.spread_ratio = self.stock1_price/self.stock2_price
        
        ma_20 = []
        pressure = []
        support = []

        for idx, value in enumerate(self.spread_ratio):
            if idx < 19:
                continue
            ma_20.append(self.spread_ratio[idx-19:idx+1].mean())
            std = self.spread_ratio[idx-19:idx+1].std()
            pressure.append(ma_20[-1] + 2*std)
            support.append(ma_20[-1] - 2*std)
            
        
        label = []
        for idx, value in enumerate(self.spread_ratio[19:]):
            if value >= pressure[idx] or value <= support[idx]:
                label.append(1)
            elif abs(value-ma_20[idx])<=0.01:
                label.append(-1)
            else:
                label.append(0)
                
        dataset = []
        for i in range(19, len(self.spread_ratio)-1):
            input_data = list(self.spread_ratio[i-19:i+1])
            ans = label[(i-19)+1]
            input_data.append(ans)
            dataset.append(input_data)
            
        random.shuffle(dataset)
            
                
        return dataset
        '''
        
        self.spread_ratio = self.stock1_price/self.stock2_price
        train = self.spread_ratio
        train = pd.DataFrame(train)
        ratios_mavg5 = train.rolling(window=5, center=False).mean()
        ratios_mavg60 = train.rolling(window=60, center=False).mean()
        std_60 = train.rolling(window=60, center=False).std()

        zscore_60_5 = (ratios_mavg5 - ratios_mavg60)/std_60
        data = zscore_60_5.values
        data = np.reshape(data, (-1))
        
        data = data[59:]
        self.stock1_price = self.stock1_price[59:]
        self.stock2_price = self.stock2_price[59:]
        self.spread_ratio = self.spread_ratio[59:]
        self.cur = 19
        
        return data
        

    def reset(self):
        self.cur += 1
        return self.dataset[:self.cur]
    

    def step(self, action):
        # calculate reward
        cur_stock1_price = self.stock1_price[self.cur]
        cur_stock2_price = self.stock2_price[self.cur]
        cur_spread_ratio = self.spread_ratio[self.cur]
        
        # warning: we have known that stock1 price is always higher than stock2 price
        
        # positive action: buy x stock2, sell y stock1
        # x= cur_spread_ratio
        # y= 1
        if action == 1:
            self.agent_stock2 += cur_spread_ratio
            self.agent_stock1 -= 1
        
        # negative action: sell x stock2, buy y stock1
        # x= diffrence between 100
        # y= diffrence between 100
        elif action == -1:
            self.agent_stock2 = 100
            self.agent_stock1 = 100
        
        reward = self.agent_stock1*cur_stock1_price + self.agent_stock2*cur_stock2_price
        
            
        self.cur += 1
        
        if self.cur == len(self.dataset):
            self.cur = 20
            
        
        new_state = self.dataset[self.cur-20:self.cur]
        
        return new_state, reward, False
```
