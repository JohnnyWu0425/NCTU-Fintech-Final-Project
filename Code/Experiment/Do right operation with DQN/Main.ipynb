## System

```python
class System():
    
    def __init__(self):
        
        self.env = Environment()
        
        self.agent = Agent()
        
        
    def train(self):
        
        state = self.env.reset()
        
        total_iteration = 2300
        current_iteration = 0
        
        self.reward_history = []
        self.action_history = []
        
        while current_iteration < total_iteration:
            
            state = np.reshape(state, (1, 20, 1))
            action = self.agent.sample_action(state)
            state = np.reshape(state, (20))
            new_state, daily_reward, done = self.env.step(action)
            self.agent.store_experience(state, action, new_state, daily_reward, int(done))
            
            state = new_state
            
            current_iteration += 1
            self.reward_history.append(daily_reward)
            self.action_history.append(action)
            
            print("Current Iteration {}: {}".format(str(current_iteration+1), daily_reward))
            
            self.agent.train_update_nn()
            
            if current_iteration % 5 == 0:
                self.agent.train_target_nn()
```

## Main

```python
system = System()
system.train()
```
