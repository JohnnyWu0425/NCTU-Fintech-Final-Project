## Import Package

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Load Data

```python
F_stock = pd.read_csv("../../Data/F.csv")
GM_stock = pd.read_csv("../../Data/GM.csv")
```

```python
F_stock
```

```python
GM_stock
```

```python
# extract close price
F_close = F_stock["Close"].values[:500]
GM_close = GM_stock["Close"].values[:500]
```

## Plot Stock Price Ratio

```python
# stock price
x = list(range(len(F_close)))
plt.figure(figsize=(15, 10))
plt.plot(x, F_close, color="blue", label="F Price")
plt.plot(x, GM_close, color="red", label="GM Price")
plt.legend()
plt.xlabel("Date Index")
plt.ylabel("Stock Price")
plt.show()
```

```python
GM_F_ratio = GM_close/F_close
```

```python
ma_20 = []
pressure = []
support = []

for idx, value in enumerate(GM_F_ratio):
    if idx < 20:
        continue
    ma_20.append(GM_F_ratio[idx-20:idx].mean())
    std = GM_F_ratio[idx-20:idx].std()
    pressure.append(ma_20[-1] + 2*std)
    support.append(ma_20[-1] - 2*std)
```

```python
# stock price ratio
x = list(range(len(F_close)))
plt.figure(figsize=(15, 10))
plt.plot(x[20:], GM_F_ratio[20:], color="blue", label="GM/F Ratio")
plt.plot(x[20:], ma_20, color="red", label="20MA")
plt.plot(x[20:], pressure, color="green", label="Pressure")
plt.plot(x[20:], support, color="yellow", label="Support")
plt.legend()
plt.xlabel("Date Index")
plt.ylabel("Ratio")
plt.show()
```
